# Visualization of preprocessed data  
Going to look at the data it self. 

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
scent_data = pd.read_csv("./data/preprocessed_data.csv")
scent_data.head()
```

```python
scent_data.shape
```

```python
# Making the memort usage small (Preventing Memory Error when changing the scale of the columns.)
# float64 -> float32, object -> category
target_columns = ["ZnO (1)", "ZnO (2)", "ZnO (3)", "SnO2 (1)", "SnO2 (2)", "SnO2 (3)", "AZO (1)", "AZO (2)", "AZO (3)"]
scent_data[target_columns + ["Time (sec)"]] = scent_data[target_columns + ["Time (sec)"]].astype("float32")
scent_data[["scent_name", "on_off_state"]] = scent_data[["scent_name", "on_off_state"]].astype("category")
```

```python
scent_data.info()
```

```python
# Since the scale of the current is too small (10^(-10)) let's chage the column.
scent_data[target_columns] = scent_data[target_columns].apply(lambda x: (10**10)*x)
scent_data.head()
```

Now the curent's unit is [10^-10 A]

```python
sns.set(rc = {'figure.figsize':(15,24)})
fig, ((ax11, ax12, ax13), (ax21, ax22, ax23), (ax31, ax32, ax33)) = plt.subplots(nrows= 3, ncols=3)
sns.violinplot(x = "ZnO (1)" ,y = "scent_name",data = scent_data, ax=ax11)
sns.violinplot(x = "ZnO (2)" ,y = "scent_name",data = scent_data, ax=ax12)
sns.violinplot(x = "ZnO (3)" ,y = "scent_name",data = scent_data, ax=ax13)
sns.violinplot(x = "SnO2 (1)" ,y = "scent_name",data = scent_data, ax=ax21)
sns.violinplot(x = "SnO2 (3)" ,y = "scent_name",data = scent_data, ax=ax22)
sns.violinplot(x = "SnO2 (3)" ,y = "scent_name",data = scent_data, ax=ax23)
sns.violinplot(x = "AZO (1)" ,y = "scent_name",data = scent_data, ax=ax31)
sns.violinplot(x = "AZO (2)" ,y = "scent_name",data = scent_data, ax=ax32)
sns.violinplot(x = "AZO (3)" ,y = "scent_name",data = scent_data, ax=ax33)
```
